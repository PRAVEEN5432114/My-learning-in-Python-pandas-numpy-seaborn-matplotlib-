```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Store': ['Store_A', 'Store_B'],
    'Jan_Sales': [100, 150],
    'Feb_Sales': [120, 130],
    'Mar_Sales': [130, 160]
})

df
```

```python
df1=df.melt(id_vars=["Store"],var_name="Month",value_name="Sales")
df1
```

```python
df2=df.melt(id_vars=["Store"],value_vars=["Jan_Sales","Feb_Sales"],var_name="Month",value_name="Sales")
df2
```

```python
df3=df.melt(id_vars=["Store"],value_vars=["Jan_Sales","Feb_Sales"],var_name="Month",value_name="Sales")
df3
```

```python
df4=df1.pivot(index="Store",columns="Month",values="Sales")
df4
```

```python
df_jan=df2.loc[df2["Month"]=="Jan_Sales"]
df_jan
```

```python
df_janpivot=df_jan.pivot(index="Store",columns="Month",values="Sales")
df_janpivot
```
