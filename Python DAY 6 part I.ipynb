```python
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
df=pd.read_excel("Sales Analysis Report.xlsx")
```

```python
df
```

```python
df.rename(columns={"Customer Name Index":"customers","Region Index":"regions","Product  Index":"products","Order Quantity":"orders","Total Unit Cost":"unit cost","Total Revenue":"revenue"},inplace=True)
```

```python
df.head()
```

```python
pd.to_datetime(df["OrderDate"])
df["YEAR"]=df["OrderDate"].dt.year
df["MONTH"]=df["OrderDate"].dt.month_name()
df["DAYS"]=df["OrderDate"].dt.day_name()
```

```python
df.head()
```

```python
df["weekday"]=df["OrderDate"].dt.weekday
```

```python
df.head()
```

```python
df["WEEKEND"]=df["weekday"].apply(lambda x: "Weekend" if x>=5 else "weekdays")
```

```python
df
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
sb.barplot(x="Channel",y="revenue",data=df)
plt.xlabel("TYPES OF CHANNEL")
plt.ylabel("REVENUE OF EACH CHANNEL")
plt.show()
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
sb.barplot(x="Channel",y="revenue",data=df,estimator=sum)
plt.xlabel("TYPES OF CHANNEL")
plt.ylabel("REVENUE OF EACH CHANNEL")
plt.show()
```

```python

plt.figure(figsize=(12,6))
sb.barplot(x="MONTH",y="orders",data=df,estimator=sum)
plt.xlabel("ORDERED MONTH")
plt.ylabel("QUANTITY OF ORDERS")
plt.show()
```

```python
df_line=df.groupby("YEAR")["revenue"].sum()
Unique_years=df_line.index
plt.figure(figsize=(12,6))
df_line.plot(kind='line',marker='o',color='red', linestyle='--')
plt.xticks(Unique_years)
plt.show()

```

```python
df_line=df.groupby("YEAR")["revenue"].sum()
plt.figure(figsize=(100,6))
df_line.plot(kind='pie',labels=df_line.index,autopct="%1.1f%%")
plt.show()
```
